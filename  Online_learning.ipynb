# Online Learning

```python
import numpy as np
from sklearn import linear_model #sklearn module provide algorithm called linear regressor and schotise linear regressor
```

```python
n_samples,n_features=1,500
y=np.random.randn(n_samples)
X=np.random.randn(n_samples,n_features)
X
```

we have generated 500 features in one single row

```python
clf=linear_model.SGDRegressor()
type(clf)
```

SGDRegressor() along with fit method have partial fit method in which we can train our model giving small chunck of data and at point we can contiue our training next time

```python
import time
```

```python
start_time=time.time()
clf.fit(X,y)
elapsed_time=time.time()-start_time
print("start_time :",start_time)
print("elapsed time :",elapsed_time)
```
