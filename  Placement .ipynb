```python
# we will be extracing input and output coulumn features are independent
X=df.iloc[:,0:2] #features
y=df.iloc[:,2:]  #labels
y2=y
y2
y=y2
y
```

```python
# lets split the data
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
 
 
```

```python
#lets scale this into -1 to 1
from sklearn.preprocessing import StandardScaler #this is a class
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train) #first it will understand the data then transform
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
y_train.values.ravel()
```

```python
clf.fit(X_train,y_train.values.ravel())
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# from mlxtend.plotting import plot_decision_regions
import pickle
```

```python
pickle.dump(clf,open('model.pkl','wb'))
```
