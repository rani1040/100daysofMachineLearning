```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,y_train,y_test=train_test_split(df.drop('placed',axis=1),df['placed'],test_size=.3,random_state=42)
```

```python
X_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_scaled_train=scaler.transform(X_train)
x_scaled_test=scaler.transform(x_test)
```

```python
X_scaled_train=pd.DataFrame(X_scaled_train,columns=X_train.columns)
x_scaled_test=pd.DataFrame(x_scaled_test,columns=x_test.columns)
```
