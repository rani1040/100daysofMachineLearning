## *****   WebScrapping


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
 
```

We can pass header into requests.get function so that websites can distinguish between bot and human, becauses they have created list of headers to access denied

```python
webpage=requests.get("https://www.ambitionbox.com/list-of-companies?page=1").text
webpage #it is same when you right click your website -->view page source 
```

```python
# 3lets give webpage to variable 
# xml is parser to travel easily
soup=BeautifulSoup(webpage,'lxml')
```

```python
# print(soup.prettify()) #format code --we can understand concept
```

```python
company=soup.find_all('div',class_='company-content-wrapper')
```

```python
len(company)
```

```python
company_name=[]
rating=[]
company_type=[]
location=[]
old=[]
number_of_employee=[]
description=[]
for i in company:
    company_name.append(i.find('h2').text.strip())
    rating.append(i.find_all('p',class_='rating')[0].text.strip())
    company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    location.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    number_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    description.append(i.find_all('p',class_="description")[0].text.strip())
     
    
```

```python
df={
    "Name":company_name,
    "Rating":rating,
     "Type":company_type,
     "HeadQuarter":location,
     "old":old,
    "Number_of_Employee":number_of_employee,
    "Description":description
    
}
```

```python
df=pd.DataFrame(df)
```
