## PipeLine 
**Pipelines chains together multiple steps so that the output of each step is used as input to the next step.**

**Pipelines makes it easy to apply the same preprocessing to train and test.**

```python
# Without Pipeline
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("ttrain.csv")
df.sample(4)
```

```python
# we dont find this column helpfull so we are dropping it 
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df
```

### Step 1 : Split data 

```python
X_train, x_test , y_train, y_test = train_test_split(df.drop(columns=['Survived']),
                                                    df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.isnull().sum()
```

## Missing value : Imputer  

```python
sa = SimpleImputer()
x1= sa.fit_transform(X_train[['Age']])
y1 = sa.transform(x_test[['Age']])

se = SimpleImputer(strategy= 'most_frequent')
x2=   se.fit_transform(X_train[['Embarked']])
y2 = se.transform(x_test[['Embarked']])

x1
x2
```

## One Hot Encoding 

```python
ohe_sex = OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore') #if found new value in future ignore it 
x_train_sex= ohe_sex.fit_transform(X_train[['Sex']])
x_test_sex= ohe_sex.transform(x_test[['Sex']])


ohe_embarked = OneHotEncoder(sparse=False,handle_unknown='ignore')
x_train_emb = ohe_embarked.fit_transform(X_train[['Embarked']])
x_test_emb = ohe_embarked.transform(x_test[['Embarked']])
```

```python
x_train_sex
x_test_emb 
```

## Concatenate 

```python
X_train_remening = X_train.drop(columns = ['Sex','Embarked','Age'])
X_test_remening  = x_test.drop(columns = ['Sex','Embarked','Age'] )
```

```python
X_train_transformed = np.concatenate((X_train_remening,x_train_emb,x_train_sex,x1),axis=1)
X_train_transformed
```

```python
x_test_transformed = np.concatenate((X_test_remening,x_test_emb,x_test_sex,y1),axis=1)
```

## Model_training 

```python
clf =DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred =clf.predict(x_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
