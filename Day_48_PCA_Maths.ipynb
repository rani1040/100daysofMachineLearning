```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(23)
```

```python
mu_vec1=np.array([0,0,0])
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]]) #identity matrix
class1_sample=np.random.multivariate_normal(mu_vec1,cov_mat1,20) # 20 rows having value between -2 and 2
```

```python
df=pd.DataFrame(class1_sample,columns=['feature1','feature2','feature3'])
df['target']=1
df
```

```python
mu_vec2=np.array([1,1,1])
cov_mat2=np.array([[1,0,0],[0,1,0],[0,0,1]]) #identity matrix
class2_sample=np.random.multivariate_normal(mu_vec2,cov_mat2,20) # 20 rows having value between -2 and 2

df2=pd.DataFrame(class2_sample,columns=['feature1','feature2','feature3'])
df2['target']=0
df2
```

```python
df=df.append(df2,ignore_index=True)
```

```python
df
```

plotly creates 3 dimension plot

```python
import plotly.express as px
fig=px.scatter_3d(x=df['feature1'],y=df['feature2'],z=df['feature3'],color=df['target'].astype(str))
fig.update_traces(marker=dict(size=12,line=dict(width=2,
                                               color='DarkSlateGrey')),
                 selector=dict(mode='markers'))
fig.show()
```

## Step 1: Mean Centering

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df.iloc[:,:3]=scaler.fit_transform(df.iloc[:,0:3])
```

```python
df.iloc[:,:3].head()
```

## Step 2 : Covariance Metrics

```python
covariance_matrix=np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
covariance_matrix # diagonal elements are : variances and non diagonal is covariance
```

## Step 3 : Finding Eigen values and eigen vectors

```python
eigen_val,eigen_vec=np.linalg.eig(covariance_matrix) # 3 dimension - 3 eigen value
```

```python
eigen_val
```

```python
eigen_vec
```

### ploting this vector in 3 d space

```python
%matplotlib inline

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from matplotlib.patches import FancyArrowPatch


class Arrow3D(FancyArrowPatch):
    def __init__(self,xs,ys,zs,*args,**kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs
        
    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
        
fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(111,projection='3d')
ax.plot(df['feature1'],df['feature2'],df['feature3'],'o',markersize=8,color='red', alpha=0.5)
ax.plot([df['feature1'].mean()],[df['feature2'].mean()],[df['feature3'].mean()],'o', markersize=15, color='red', alpha=0.5)   

for v in eigen_vec.T:
    a=Arrow3D([df['feature1'].mean(),v[0]],[df['feature2'].mean(),v[1]],[df['feature3'].mean(),v[2]],mutation_scale=20, lw=3, arrowstyle="-|>",color='blue')
    ax.add_artist(a)

ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_zlabel('z_values')

plt.title('Eigenvectors')

plt.show()
```

# x and y direction vector 

```python
pc=eigen_vec[0:2]
pc
```
